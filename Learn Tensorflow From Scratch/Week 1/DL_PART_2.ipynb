Let's try to predict the price of a house in the most simple way possible using neural network. Let's assume that a house has a base price of 50k and with every additional bedroom, a cost of 50k is added. <br/>
That is the relationship is like - house_price = 50 + house x 50 <br/>
So for a 7 bedroom house, the price is 50+7x50=400. Now we will use the model to try and predict this price.

```python
#Importing libraries
import tensorflow as tf
import numpy as np
```

We will try creating dataset first using the method that we have previously seen, but this time we will use function. We will have 2 one dimensional tensors - 1 dimensional tensor to define the number of bedrooms in a house and 1 dimensional tensor to define the prices of the houses.

```python
def create_dataset():
    bedrooms = np.array([1,2,3,4,5,6], dtype=float)  # Calling data type as float for satisfying multiple conditions like half room
    prices = np.array([1,1.5,2,2.5,3,3.5], dtype=float)  # prices are in hundreds of thoudands dollars
    return bedrooms, prices
```

```python
#bedroom numbers are our features and prices of house are our target
features, target = create_dataset()

print('Features shape - ', features.shape)
print('Target shape - ', target.shape)
```

Since we have our data ready, we can define and compile the model now. We will use a function for the same.

```python
def define_and_compile_model():
    model = tf.keras.models.Sequential([
        tf.keras.Input(shape=(1,)),  #As we have seen the input is of 1D shape
        tf.keras.layers.Dense(units=1)  # 1 layer with 1 neuron
    ])
    model.compile(optimizer='sgd', loss='mean_squared_error')
    return model
```

```python
untrained_model = define_and_compile_model()
untrained_model.summary()
```

Summary looks similar to the previous lab code. Now we have to train the model with our data using the fit method. For this as well, we will use function

```python
def train_model():
    bedrooms, prices = create_dataset()  # This is calling the dataset and storing here directly
    model = define_and_compile_model()  # Or you could have used untrained_model , they are the same
    model.fit(bedrooms, prices, epochs=500)
    return model
```

```python
trained_model = train_model()
```

Once the training is done, you can predict new data using the model -

```python
new_data = np.array([7.0], dtype=float)
new_price = trained_model.predict(new_data)
new_price
```

```python
new_price.item()
```

So the model predicted quiet well in this case. We are expecting it to be close to 4, if not exactly 4.
